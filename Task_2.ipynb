```python
#problem 3

if __name__ == '__main__':
    stud = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        stud.append([name,score]) 
    x = 99
    for i in range(len(stud)):
        if x > float(stud[i][1]):
            x = float(stud[i][1])
    y = 99
    for i in range(len(stud)):
        if float(stud[i][1]) > float(x) and y > float(stud[i][1]):
            y = float(stud[i][1])
    runner_up = []
    for i in range(len(stud)):
        if float(stud[i][1]) == float(y):
            runner_up.append(stud[i][0])
    runner_up = sorted(runner_up)

    for i in range(len(runner_up)):
        print(runner_up[i])
```

```python
#problem 4 

if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    
    if query_name in student_marks:
        x = ((float(student_marks[query_name][0]) + float(student_marks[query_name][1]) + float(student_marks[query_name][2])) / 3)
    
    print('%.2f' % x)
```

```python
#problem 5 

if __name__ == '__main__':
    N = int(input())
    empty_list = []
    conv = []

    for i in range(N):
        x = input().split()
        empty_list.append(x)

    for i in range(len(empty_list)):
        if empty_list[i][0] == 'insert':
            x = int(empty_list[i][1])
            y = int(empty_list[i][2])
            conv.insert(x,y)
        elif empty_list[i][0] == 'print':
            print(conv)
        elif empty_list[i][0] == 'remove':
            conv.remove(int(empty_list[i][1]))
        elif empty_list[i][0] == 'append':
            conv.append(int(empty_list[i][1]))
        elif empty_list[i][0] == 'sort':
            conv.sort()
        elif empty_list[i][0] == 'pop':
            conv.pop()
        elif empty_list[i][0] == 'reverse':
            conv.reverse()
```
